# Direct Sampling

Direct Sampling is statistical method to get approximation.  

Pi(π) approximation is obtained by using this method.

If we think the most simple case, we throw the ball into the square area (x coordinate from -1 to 1 and y coordinate from -1 to 1).

We repeat throwing the ball into **Square** many times and count the number of ball that is in **Circle**.

\# of ball in Circle / # of ball in Square 
= Area of Circle / Area of Square  
= (pi * radius^2) / (2*2)

Thus, pi = 4 × radius × (radius^2) × (# of ball in Circle / # of ball in Square)

In this way, we can get pi approximation.



##Function

```python
import random, math
import numpy as np
import matplotlib.pyplot as plt


def pi_value(number):
    
    in_circle_number = 0
    data_x = []
    data_y = []

    # Repeat throwing particles
    for i in range(number):
        x = random.uniform(-radius, radius)
        y = random.uniform(-radius, radius)
        
        
        distance = math.sqrt(x**2 + y**2)
        #If the particle is inside circle, add 1 to in_circle_number.
        if distance < radius:

            in_circle_number += 1

        data_x.append(x)
        data_y.append(y)
    
    return in_circle_number, data_x, data_y

number = 10000
radius = 1

pi_temp, result_x, result_y = pi_value(number)  


pi = 4 * pi_temp / (number)

print('pi value: ', pi)
```

##Graph

```python
color = []

# If the particle is in circle, color of particle is green. 
# Otherwiese, if the particle is out of circle, color of particle is blue.

for i in range(number):
    distance = math.sqrt(result_x[i]**2 + result_y[i]**2)
    if distance < radius: 
        color.append('g')
    else:
        color.append('b')        

circle = plt.Circle((0,0), radius, color='r', fill=False, linewidth=3)
flg, ax = plt.subplots()
ax.add_patch(circle)

plt.scatter(result_x, result_y, c=color, s=20)

plt.text(-radius, radius, ('inside circle: '+ str(pi_temp)+' / '+str(number)))

plt.show()
```
