# Markov Chain

Estimate Pi value by using Markov Chain Model.

Markov Chain Model is that present situation is only depended on last past situation.

In aspect of getting pi, Markov Chain Model is same with Direct Sampling, but Direct Sampling is different from Markov Chain Model.

In this model, initial coordinate is fixed; x = -1, y= -1
We have to thnink the square that x coordinate is from -1 to 1 and y coordinate is from -1 to 1.

We get the two random number between -1 and 1, and add them to initial coordinates each other. If the new coordinate is inside the square, new coordinate is next coordinate. Otherwise(not inside the square), last coordinate (before adding) is next coordinate.

We save all coordinates as matrix, and count the number only when the distance of coordinate from origin is less than 1.

pi = 4 * (counted number) / (total trial)

And repeat this process 5 times and save pi values as table using pandas

```python
import math, random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def throwing(number):
    #Initial Coordinate
    x = -1
    y = -1

    coordinates = []
    for i in range(number):
        x_add = random.uniform(-1, 1)
        y_add = random.uniform(-1, 1)

        if abs(x + x_add) < 1 and abs(y + y_add) < 1:
            x = x + x_add
            y = y + y_add
        else:
            pass

        coordinates.append([x,y])
    return coordinates

def counting(coordinates):
    count_number = 0
    
    for i in coordinates:
        x, y = i
        distance = math.sqrt(x**2 + y**2)
        if distance < 1:
            count_number += 1
    
    return count_number

result = []
number = 10000 # number of tryiing

for i in range(5): 
    result_matrix = throwing(number)   
    number_hit = counting(result_matrix)
    pi = 4 * number_hit / number
    result.append(pi)

index = ["1", "2", "3", "4", "5"]
column = ["pi"]

pd.DataFrame(result, index, column)
```

## Graph

```python
# graph
result_matrix = throwing(number)
x_graph = []
y_graph=[]

for i in result_matrix:
    x, y = i
    x_graph.append(x)
    y_graph.append(y)

circle = plt.Circle((0,0), 1, color='r', fill=False, linewidth=1)
flg, ax = plt.subplots()
ax.add_patch(circle)

plt.plot(x_graph,y_graph) 
```
