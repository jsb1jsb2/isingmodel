#Ising Model 2d



##import python function


'n' is matrix size (n by n matrix) 


```python
import random, math
import matplotlib.pyplot as plt
import numpy as np

global n, J, N

n = 10
J = 1
```

## Functions

1. Make n by n random matrix (composed of -1 and 1 )

```python
def mk_matrix_2d():
    data_init = np.zeros([n,n])
    data_init = np.random.choice([-1,1], size = (n,n))
    return data_init
```

2. Determine whether to flip 

    Get the random xy coordinate and Determine whether to flip that coordinate by comparing pi value

    beta = 1 / temperature

```python
def pi_value_2d(data_unit, beta):

    Energy = 0

    x = random.randint(0,(n-1))
    y = random.randint(0,(n-1))

    top = data_unit[(y-1)%n, x]
    bottom = data_unit[(y+1)%n, x]
    right = data_unit[y, (x+1)%n]
    left = data_unit[y, (x-1)%n]
    Energy += data_unit[y, x] * (top + bottom + right + left)

    pi_a = math.exp((+1)*beta*Energy)
    pi_b = math.exp((-1)*beta*Energy)
    pi_divided = pi_b / pi_a

    if pi_divided <= 1:
        if random.random() < pi_divided:
            data_unit[y,x] = -data_unit[y,x]
    elif pi_divided > 1:
        data_unit[y,x] = -data_unit[y,x]

    return data_unit
```

3. Repeat determining whether to flip 

    for making randomized matrix

    beta = 1/0.1



```python
def matrix_repeat_2d(number, matrix_matrix):
    for i in range(number):
        matrix_matrix = pi_value_2d(matrix_matrix, 1/0.1)
    return matrix_matrix
```

4. Calculate Average Energy, Magnetization, Heat Capacity, Susceptibility
    
    in certain temperature

```python
def cal_temperature(matrix_cal, beta):

    Sum_Energy = 0
    Sum_Energy_square = 0
    Avg_Energy = 0
    Avg_Energy_square = 0

    Sum_M_square = 0
    Sum_M = 0
    Avg_M=0
    Avg_M_square = 0

    Sum_Hc = 0
    Sum_Sus = 0

    #Repeat N times

    for j in range(N):

        H = 0
        H_square = 0

        #Calculate Hamiltonian
        
        for x in range(n):
            for y in range(n):
                top = matrix_cal[(y-1)%n, x]
                bottom = matrix_cal[(y+1)%n, x]
                right = matrix_cal[y, (x+1)%n]
                left = matrix_cal[y, (x-1)%n]

                h_temp =  -J * (matrix_cal[y, x] * (top + bottom + right + left)) / 2
                
                H += h_temp / (n*n)
                H_square += h_temp**2 / (n*n)

        Sum_Hc += beta * beta * (H_square - H*H)
        
        M = np.sum(matrix_cal)

        Sum_Energy += H
        Sum_Energy_square += H_square
        Sum_M += abs(M)
        Sum_M_square += M*M
        matrix_cal = pi_value_2d(matrix_cal, beta)

    #Get average values by dividing N

    Avg_Energy = Sum_Energy / N
    Avg_Energy_square = Sum_Energy_square / N

    Avg_M = Sum_M / N
    Avg_M_square = Sum_M_square / N

    Avg_Hc = Sum_Hc / N

    Sus = beta * ((Avg_M_square) - Avg_M*Avg_M) /(n*n)

    return Avg_Energy, Avg_M, Avg_Hc, Sus
```

## Main Code



1. Make random matrix and Randomize

    N is repeat number
    
    T is temperature matrix (from 1.0K to 5.0K with 0.01 interval)

```python
init = mk_matrix_2d()
matrix_cel = matrix_repeat_2d(5000, init)

N = 10000

#T = np.arage(start temperature, end temperature, interval)
T = np.arange(0.9, 5.1, step=0.01)

```

2. Calculate Average Energy, Heat Capaticy, Magnetizaion, Susceptibility

    from 1.0K to 5.0K

```python
result_Avg_Energy = []
result_Hc = []
result_Sus = []
result_Mg =[]

for i in range(len(T)):
    temp1, temp2, temp3, temp4 = cal_temperature(matrix_cel, 1/T[i])
    result_Avg_Energy.append(temp1)
    result_Mg.append(temp2)
    result_Hc.append(temp3)
    result_Sus.append(temp4)
```

3. Graph

```python
j = 0
y_label = ['Average Energy', 'Susceptibility', 'Heat Capacity', 'Magnetization']
for i in [result_Avg_Energy, result_Sus, result_Hc, result_Mg]:
    plt.scatter(T, i)
    plt.xlabel('Temperature')
    plt.ylabel(y_label[j])
    plt.show()
    j += 1
```
